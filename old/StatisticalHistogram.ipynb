```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#subject = subject number
#type = VR or Non-VR
def loadSubjectData(subject, type):
    path = "Pre-Processed Data/subject-"+subject+"/"+type+"/Subject-"+subject+"_"+type+".csv"
    data = pd.read_csv(path)
    data = pd.DataFrame(data)
    return data
```

```python
#Change subject, VR/Non-VR, and channel here
subject = 17
nonVR = loadSubjectData(str(subject),"Non-VR")
VR = loadSubjectData(str(subject),"VR")
```

```python
binRanges = [None]*81
x = -200
i = 0
for i in range(len(binRanges)):
    binRanges[i]=x
    x+=5
```

### NonVR

```python
nonVR.plot.hist(bins=binRanges, figsize=[15,10], density = True)
```

```python
nonVR_mean = nonVR.mean()
nonVR_median = nonVR.median()
nonVR_std = nonVR.std()

print("Mean: ")
print(nonVR_mean)
print()
print("Median: ")
print(nonVR_median)
print()
print("Standard Deviation: ")
print(nonVR_std)
```

### VR

```python
VR.plot.hist(bins=binRanges, figsize=[15,10], density = True)
```

```python
VR_mean = VR.mean()
VR_median = VR.median()
VR_std = VR.std()

print("Mean: ")
print(VR_mean)
print()
print("Median: ")
print(VR_median)
print()
print("Standard Deviation: ")
print(VR_std)
```

# Normalized

### NonVR

```python
norm_nonVR = (nonVR-nonVR.min())/(nonVR.max()-nonVR.min())
```

```python
norm_nonVR.plot.hist(bins=50, figsize=[15,10], density = True)
```

```python
norm_nonVR_mean = norm_nonVR.mean()
norm_nonVR_median = norm_nonVR.median()
norm_nonVR_std = norm_nonVR.std()

print("Mean: ")
print(norm_nonVR_mean)
print()
print("Median: ")
print(norm_nonVR_median)
print()
print("Standard Deviation: ")
print(norm_nonVR_std)
```
